# Zadanie 1

```python
data =pd.read_csv('sample_data/president_heights (1).csv')
heights = np.array(data['height(cm)'])
print(heights)
```

```python
#1 srednia
print("Meeean height:   " , np.average(heights) )
#2 odchylenie standardowe
print("Standard deviation:  ", np.std(heights))
#3 najwyzszy oraz najnizszy wzrost
print("Minimum heeigrht:  ", np.min(heights))
print('Maximum height:' + str(np.max(heights) )   )

#5 25 i 75 kwantyl - 25th percentile
#4 mediana
print("25th percentile:  ", np.quantile(heights, 0.25))
print("Median:  ", np.median(heights))
print("75th percentile:  ", np.quantile(heights, 0.75))
```

```python
#a. Znajdź wektory własne, oraz wartości własne dla zawartej w pliku macierzy
wartosc_wlasna, wektor_wlasny = la.eig(mtx1)
print("Wektory własne macierzy:", wektor_wlasny)   
print("Wartości własne macierzy:",  la.eigvals(mtx1) )   
#b. Oblicz macierz odwrotną dla macierzy z pliku
print("Macierz odwrotna:  ",  np.invert(mtx1)  )

#macierz transponowana
# np.transpose(mat1)
```

# Zadanie 3 

```python
data3 = pd.read_csv('sample_data/Seattle2014 (1).csv', delimiter=',')
data3
rainfall = pd.read_csv('sample_data/Seattle2014 (1).csv')['PRCP'].values
inches = rainfall/254.0 

import matplotlib.pyplot as plt
import seaborn; seaborn.set() 
plt.hist(inches, 40 )
```

```python
#dni gdy nie padało 
dni_bz_opadow = data3.DATE[data3.PRCP == 0]
print("dni_bz_opadow:  ",  dni_bz_opadow )

#dni gdy padało
dni_z_opadami = data3.DATE[data3.PRCP > 0]
print("dni_z_opadami:  ",  dni_z_opadami  )

data3 = pd.DataFrame(data3)
data3['inches'] = inches
#Spadło powyżej 0.5 cali (inch) deszczu
#print(data3)
dni_pow_05 = data3.DATE[data3.inches > 0.5]
print("dni_z_opadami_pow_05:  ",  dni_pow_05  )
dni_pow_05.count
#Spadło poniżej 0.2 cali (inch) deszczu, ale padało!
dni_pon_02 = data3.DATE[(data3.inches < 0.2) & (data3.PRCP > 0)]
print("dni_z_opadami_pon_0.2:  ",  dni_pon_02  )
```

```python
#Korzystając z maskowania policz następujące statystyki:
#Medianę opadów w deszczowe dni w 2014 roku
import datetime
data3['date'] = pd.to_datetime(data3['DATE'])

opady_w_deszczowe_dni = data3.PRCP[data3.PRCP > 0]
print("Medianę opadów w deszczowe dni w 2014 roku:  ",  np.median(opady_w_deszczowe_dni)  )

#Medianę opadów latem w 2014 roku (czyli dni pomiędzy dniem 172 a 262)
lato = np.array(data3.PRCP[171:262])
print("Medianę opadów w deszczowe dni w 2014 roku:  ",  np.median(lato)  )
#Maksymalne opady latem 2014 roku
print("Max opadów w deszczowe dni w 2014 roku:  ",  np.max(lato)  )
#Maksymalne opady poza latem 2014 roku (czyli wiosna, jesień i zima)
nie_lato1 = np.array(data3.PRCP[0:171])
nie_lato2 = np.array(data3.PRCP[262:])
nie_lato = np.concatenate((nie_lato1, nie_lato2))
print("Max opadów w deszczowe dni w 2014 roku:  ",  np.max(nie_lato)  )
```
